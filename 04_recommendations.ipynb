```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Custom palette
COLORS = ['#F0B429', '#C87D0E', '#D81B7A', '#7B2D8B', '#9B8BB4',
          '#E85D9A', '#5B1F7A', '#F7D26A', '#B5478A', '#C9A0DC']
sns.set_palette(COLORS)
sns.set_theme(style="whitegrid")

# Load Power BI dataset (has all our engineered features)
df = pd.read_csv('C:/churn_project/outputs/powerbi_data.csv')

print("Data loaded ✅")
print("Shape:", df.shape)
print("\nSegment counts:")
print(df['Segment'].value_counts())
```

```python
# Deep dive into each segment
print("=" * 60)
print("         SEGMENT DEEP DIVE ANALYSIS")
print("=" * 60)

for segment in ['Critical', 'Loyal', 'At Risk', 'Safe']:
    seg_df = df[df['Segment'] == segment]
    
    print(f"\n{'='*60}")
    print(f"  {segment} Segment ({len(seg_df)} customers)")
    print(f"{'='*60}")
    
    print(f"\n  💰 Financial:")
    print(f"     Avg CLV              : ₹{seg_df['CLV'].mean():,.2f}")
    print(f"     Avg Monthly Charges  : ₹{seg_df['MonthlyCharges'].mean():,.2f}")
    print(f"     Avg Churn Probability: {seg_df['ChurnProbability'].mean():.2%}")
    print(f"     Total Revenue at Risk: ₹{seg_df['RevenueAtRisk'].sum():,.2f}")
    
    print(f"\n  📋 Contract Type:")
    contract_pct = seg_df['Contract'].value_counts(normalize=True) * 100
    for contract, pct in contract_pct.items():
        print(f"     {contract:<20}: {pct:.1f}%")
    
    print(f"\n  ⏱️ Tenure Group:")
    tenure_pct = seg_df['TenureGroup'].value_counts(normalize=True) * 100
    for tenure, pct in tenure_pct.items():
        print(f"     {tenure:<20}: {pct:.1f}%")
    
    print(f"\n  💳 Payment Method:")
    payment_pct = seg_df['PaymentMethod'].value_counts(normalize=True) * 100
    for payment, pct in payment_pct.items():
        print(f"     {payment:<20}: {pct:.1f}%")

print("\n" + "=" * 60)
print("Deep dive complete ✅")
```
